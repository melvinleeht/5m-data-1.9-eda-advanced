```python
tips = pd.read_csv("data/tips.csv")

tips
```

```python
tips["tip_pct"] = tips["tip"] / tips["total_bill"]
```

```python
tips
```

```python
def top5(df, n=5, column="tip_pct"):
    return df.sort_values(column, ascending=False)[:n]
```

```python
tips.agg(top5)
```

```python
top5(tips, n=6)
```

```python
def top5_series(s):
    return s.sort_values(ascending=False).head(5)
```

```python
tips["tip_pct"].agg(top5)
```

```python
tips.agg(top5_series)
```
